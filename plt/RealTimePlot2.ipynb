```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from time import sleep

m = 100
n = 100
matrix = np.random.normal(0,1,m*n).reshape(m,n)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(0, 10):
    ax.clear()
    ax.plot(matrix[i,:])
    fig.canvas.draw()
    sleep( 0.1 )
```
