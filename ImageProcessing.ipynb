```python
import matplotlib.pyplot as plt, numpy as np
from skimage import data
from skimage.transform import resize
from PIL import Image 
```

```python
image = data.camera()
print( image.shape )
plt.imshow( image, cmap='gray' )
plt.show()
image = resize(image, ( 10, 10))
image = np.where( image > 0.5, 1, 0 )
plt.imshow( image, cmap='gray' )
plt.show()
image = resize(image, ( 512, 512), preserve_range=False, anti_aliasing=True)
image = np.where( image > np.mean(image), 1, 0 )
plt.imshow( image, cmap='gray' )
plt.show()
```
