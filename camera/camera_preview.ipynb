```python
from time import sleep
import picamera

with picamera.PiCamera() as camera:
    camera.start_preview()
    camera.hflip = True
    camera.vflip = True
    sleep(2)
    camera.capture('image.jpg') 
    camera.stop_preview()
    print("Capture completed.")
```
