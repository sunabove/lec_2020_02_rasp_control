```python
from time import sleep
import picamera

with picamera.PiCamera() as camera:
    camera.start_preview()
    camera.hflip = False
    camera.vflip = False
    sleep(2)
    camera.capture('image.jpg') 
    camera.stop_preview()
    print("Capture completed.")
pass
```
